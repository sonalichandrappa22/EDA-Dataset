```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python

df = pd.read_excel("/content/Electric_Vehicle_Population_Data.xlsx")
```

```python
#How many electric vehicles are recorded in the dataset?
df.head()
```

```python
# How many unique EV manufacturers (Make) are present in the dataset?
df.info()
```

```python
# Which manufacturer has the highest number of registered electric vehicles?
df.shape
```

```python
# What are the top 10 EV manufacturers based on the number of vehicles?
df.columns
```

```python
#How many Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs) are there?
df.drop_duplicates()
```

```python
#What percentage of the total vehicles are BEVs and PHEVs?
print(df.isnull().sum())
```

```python
# What is the average electric range of all vehicles in the dataset?
avg_range = df['Electric Range'].mean()
avg_range

```

```python
# Which vehicle has the maximum electric range, and what is that range
max_range_vehicle = df.loc[df['Electric Range'].idxmax()]
max_range_vehicle
```

```python
#What are the top 10 cities with the highest EV population
top_10_cities = df['City'].value_counts().head()
top_10_cities
```

```python
#Which county has the highest number of registered electric vehicles
top_County = df['County'].value_counts().head(10)
top_County
```

```python
#Which state has the highest number of electric vehicles
top_State = df['State'].value_counts().head()
top_State
```

```python
#What are the top 10 most popular EV models
top_10_models = df['Model'].value_counts().head(10)
top_10_models
```

```python
#Which EV model is the most frequently registered
EV_Model_frequently_registered = df['Model'].value_counts().mode()
EV_Model_frequently_registered
```

```python
#How has EV adoption changed over the years based on the Model Year column
EV_adoption_by_year = df['Model Year'].value_counts().sort_index()
EV_adoption_by_year

```

```python
#How has EV adoption changed over the years based on the Model Year column
ev_adoption = df['Model Year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.plot(ev_adoption.index, ev_adoption.values, marker='o')
plt.title('EV Adoption Over the Years')
plt.xlabel('Model Year')
plt.ylabel('Number of EV Registrations')
plt.grid(True)
plt.show()
```

```python
#Which model year recorded the highest number of EV registrations
highest_ev_year = df['Model'].value_counts().idxmax()
highest_ev_year
```

```python
#What is the average electric range for each vehicle manufacturer
avg_range_by_make = df.groupby('Make')['Electric Range'].mean()
avg_range_by_make
```

```python
# Which manufacturer offers the highest average electric range
highest_avg_range_make = df.groupby('Make')['Electric Range'].mean().idxmax()
highest_avg_range_make
```

```python
#What is the distribution of Clean Alternative Fuel Vehicle (CAFV) eligibility among vehicles
CAFV_distribution = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
CAFV_distribution
```

```python
# Is there a relationship between Base MSRP and Electric Range
df = df[['Base MSRP', 'Electric Range']].dropna()
df
```

```python
#What is the average electric range of BEVs compared to PHEVs
avg_range = df.groupby("Electric Vehicle Type")["Electric Range"].mean()
avg_range
```
